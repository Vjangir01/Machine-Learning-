```python
# here we take averge value of '-' seperated value by divided by 2
```

```python
def process(r):
    if '-' in r:
        result = (float(r.split('-')[0])+float(r.split('-')[1]))/2
        return result
    else:
        return r
```

```python
X['new_total_sqft'] = df['total_sqft'].apply(process)
```

```python
X['new_total_sqft'] = X['new_total_sqft'].astype('float')
```

```python
import re
```

```python
def fetch_number(r):
    result = re.findall(r'\d+' ,str(r))
    return result[0]
# seperate int value from string
```
